```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv("data.csv")  # Make sure "data.csv" is in your working directory

# Set visualization style
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# Assuming data is already loaded in variable `data`
# Display basic info
print("Dataset Shape:", data.shape)
print("\nColumn Types:\n", data.dtypes)

# Missing values check
print("\nMissing Values:\n", data.isnull().sum())

# Attrition Rate
attrition_rate = data['Attrition'].value_counts(normalize=True)['Yes'] * 100
print(f"\nAttrition Rate: {attrition_rate:.2f}%")

# Countplot of Attrition
sns.countplot(data=data, x='Attrition', palette='Set2')
plt.title('Attrition Count')
plt.show()

# Attrition vs Age
sns.boxplot(data=data, x='Attrition', y='Age', palette='Set3')
plt.title('Age vs Attrition')
plt.show()

# Attrition vs YearsAtCompany
sns.boxplot(data=data, x='Attrition', y='YearsAtCompany', palette='Set1')
plt.title('Years at Company vs Attrition')
plt.show()

# Attrition vs MonthlyIncome
sns.boxplot(data=data, x='Attrition', y='MonthlyIncome', palette='Set2')
plt.title('Monthly Income vs Attrition')
plt.show()

# Correlation Heatmap for numeric variables
numeric_cols = data.select_dtypes(include=np.number).columns
plt.figure(figsize=(14, 10))
corr_matrix = data[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# Age distribution with Attrition
sns.histplot(data=data, x='Age', hue='Attrition', multiple='stack', kde=True, palette='Set1')
plt.title('Age Distribution by Attrition')
plt.show()

# YearsAtCompany distribution
sns.histplot(data=data, x='YearsAtCompany', hue='Attrition', multiple='stack', kde=True, palette='Set3')
plt.title('Years at Company by Attrition')
plt.show()

# MonthlyIncome distribution
sns.histplot(data=data, x='MonthlyIncome', hue='Attrition', multiple='stack', kde=True, palette='Set2')
plt.title('Monthly Income by Attrition')
plt.show()

# Summary conclusions
print("\nKey Insights:")
print("- Attrition rate is {:.2f}%.".format(attrition_rate))
print("- Younger employees and those with fewer years at the company are more likely to leave.")
print("- Lower income seems to correlate with higher attrition.")
```
